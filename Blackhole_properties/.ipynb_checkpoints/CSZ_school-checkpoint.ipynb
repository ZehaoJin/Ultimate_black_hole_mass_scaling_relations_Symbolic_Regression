```python
df = pd.read_csv('SMBH_Data_0522.csv',header=1)
```

```python
nanpercent=0.20
cols_to_delete = df.columns[df.isnull().sum()/len(df) > nanpercent]
df=df.drop(cols_to_delete, axis = 1)
df=df.dropna(axis='index',how='any')
print(df.columns)
print(len(df.columns))
#HTML(df_purge.to_html())
```

```python
# remove stds
std_col=np.array([],dtype=int)
for i,col in enumerate(df.columns):
    if (col[-3:]=='std') & (col!='M_BH_std'):
        std_col=np.append(std_col,i)

cols=np.delete(np.arange(df.shape[1]),std_col)
df_nostd=df.iloc[:,cols]

# clean the ones that tells the answer directly
#df_clean=df_nostd.drop(['ID', 'Galaxy', 'Morphology', 'Type', 'ETG', 'Leda Type','M_BH/M*_sph', 'log_M_BH/M*_sph', 'M_BH/M*_disk', 'log_M_BH/M*_disk',
       #'M_BH/M*_gal', 'log_M_BH/M*_gal'],axis='columns')
df_clean=df_nostd.drop(['ID', 'Galaxy', 'Morphology', 'Type','T-type','ETG', 'Leda Type','M_BH/M*_sph', 'log_M_BH/M*_sph',
'M_BH/M*_gal', 'log_M_BH/M*_gal', 'M_BH_std','B/T'],axis='columns')
```
