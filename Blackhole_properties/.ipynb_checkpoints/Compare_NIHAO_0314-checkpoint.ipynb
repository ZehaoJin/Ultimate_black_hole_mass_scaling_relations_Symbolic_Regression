```python
import numpy as np
import pandas as pd
from pysr import PySRRegressor
from IPython.display import HTML
from matplotlib import pyplot as plt

from hyperfit.linfit import LinFit
```

```python
#nihao = pd.read_csv('NIHAO-BH-S-Sph-DM-BT-Sig-rho.csv')
nihao = pd.read_csv('NIHAO-BH-S-Sph-DM-BT-Sig-rho (2023-03-17).csv')
```

```python
nihao = nihao.dropna(axis='index',how='any')
nihao
```

```python
nihao_old = pd.read_csv('NIHAO-BH-S-Sph-DM-BT-Sig-rho.csv')
nihao_old = nihao_old.dropna(axis='index',how='any')
```

```python
nihao['log(sigma0)']
```

```python
nihao_old['log(sigma0)']
```

```python
obs = pd.read_csv('SMBH_Data_0303.csv',header=1)
paras=['log_sigma0','log_B/T','logRho_soi_approx_new','M_BH']
obs = obs[paras].dropna(axis='index',how='any')
obs
```

```python
#[ 2.77679962,  1.16427157, -0.36642744,  3.32060273,  0.08290356]  [0.05184539, 0.09560613, 0.037808  , 0.1058261 , 0.05875805]
def unnorm_relation(log_sigma0,log_B_T,logRho_soi):
    a=2.77679962
    b=1.16427157
    c=-0.36642744
    d=3.32060273
    return a*log_sigma0+b*log_B_T+c*logRho_soi+d
```

```python
# [ 2.85425535  1.16151575 -0.3337952   8.20260696  0.08668111] [0.03487247 0.04581018 0.02570283 0.03600486 0.05704124]
# normalization array([ 2.2771162 , -0.35953965,  2.77884623])
def norm_relation(log_sigma0,log_B_T,logRho_soi):    
    a=2.85425535
    x1norm=2.2771162
    b=1.16151575
    x2norm=-0.35953965
    c=-0.3337952
    x3norm=2.77884623
    d=8.20260696
    return a*(log_sigma0)+b*(log_B_T)+c*(logRho_soi)+d-a*x1norm-b*x2norm-c*x3norm
```

```python
#def rmse(y,y_pred,w):
#    return np.sqrt(np.average((y-y_pred)**2,weights=w))
def rmse(y,y_pred):
    return np.sqrt(np.average((y-y_pred)**2))
```

```python
plt.figure(figsize=(8,6))

# nihao
y=nihao['log(M_BH/M_Sun)'].to_numpy()
y_pred=unnorm_relation(nihao['log(sigma0)'],nihao['log(B/T)'],nihao['log(Rho_soi)']).to_numpy()
plt.scatter(y,y_pred,label='NIHAO @2 kpc',alpha=0.8)
print('NIHAO @2 kpc rmse:',rmse(y,y_pred))

# nihao 0.595 kpc
y=nihao_old['log(M_BH/M_Sun)'].to_numpy()
y_pred=unnorm_relation(nihao_old['log(sigma0)'],nihao_old['log(B/T)'],nihao_old['log(Rho_soi)']).to_numpy()
plt.scatter(y,y_pred,label='NIHAO @0.595 kpc',alpha=0.8)
print('NIHAO @0.595 kpc rmse:',rmse(y,y_pred))

# f(x)=x line
minrange=np.array([y.min(),y_pred.min()]).min()
maxrange=np.array([y.max(),y_pred.max()]).max()
plt.plot(np.linspace(minrange,maxrange),np.linspace(minrange,maxrange),label='f(x)=x',c='r',ls='--')

# observation
y=obs['M_BH'].to_numpy()
y_pred=unnorm_relation(obs['log_sigma0'],obs['log_B/T'],obs['logRho_soi_approx_new']).to_numpy()
plt.scatter(y,y_pred,label='Observations',alpha=0.8)
print('obs rmse:',rmse(y,y_pred))

plt.xlabel(r'True $\rm{log} M_{BH}$',fontsize=20)
plt.ylabel(r'Predicted $\rm{log} M_{BH}$',fontsize=20)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))

# nihao 2 kpc
y=nihao['log(M_BH/M_Sun)'].to_numpy()
y_pred=norm_relation(nihao['log(sigma0)'],nihao['log(B/T)'],nihao['log(Rho_soi)']).to_numpy()
plt.scatter(y,y_pred,label='NIHAO @2 kpc',alpha=0.8)
print('NIHAO @2 kpc rmse:',rmse(y,y_pred))

# nihao 0.595 kpc
y=nihao_old['log(M_BH/M_Sun)'].to_numpy()
y_pred=norm_relation(nihao_old['log(sigma0)'],nihao_old['log(B/T)'],nihao_old['log(Rho_soi)']).to_numpy()
plt.scatter(y,y_pred,label='NIHAO @0.595 kpc',alpha=0.8)
print('NIHAO @0.595 kpc rmse:',rmse(y,y_pred))


# f(x)=x line
minrange=np.array([y.min(),y_pred.min()]).min()
maxrange=np.array([y.max(),y_pred.max()]).max()
plt.plot(np.linspace(minrange,maxrange),np.linspace(minrange,maxrange),label='f(x)=x',c='r',ls='--')

# observation
y=obs['M_BH'].to_numpy()
y_pred=norm_relation(obs['log_sigma0'],obs['log_B/T'],obs['logRho_soi_approx_new']).to_numpy()
plt.scatter(y,y_pred,label='Observations',alpha=0.8)
print('obs rmse:',rmse(y,y_pred))

plt.xlabel(r'True $\rm{log} M_{BH}$',fontsize=20)
plt.ylabel(r'Predicted $\rm{log} M_{BH}$',fontsize=20)
plt.legend()
plt.show()
```
