```python
df=pd.read_csv('SMBH_Data_01_22_24.csv',header=1)
```

```python
# make a list of columns that contains string data
str_cols = [col for col in df.columns if df[col].dtype == object]

str_cols
```

```python
white_list=['Galaxy','Morphology','Type','Leda Type','GJC23 Type','GJC23 Dust']
```
